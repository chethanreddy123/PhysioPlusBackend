```python
import requests
from concurrent.futures import ThreadPoolExecutor

def make_request(hit_number):
    url = "https://physioplusbackend.onrender.com/SearchPatient"
    payload = {"SearchString": ""}
    headers = {"Content-Type": "application/json"}
    response = requests.post(url, json=payload, headers=headers)
    print(f"Hit {hit_number} - Response: {response.status_code}")
    # If you want to see the response data, uncomment the next line:
    # print(response.json())

# Number of simultaneous requests
num_requests = 1000

# Create a ThreadPoolExecutor with the specified number of threads
with ThreadPoolExecutor(max_workers=num_requests) as executor:
    # Submit the API requests to the thread pool
    future_to_request = {executor.submit(make_request, i+1): i for i in range(num_requests)}

    # Wait for all requests to complete
    for future in future_to_request:
        future.result()

print("All requests completed.")
```
